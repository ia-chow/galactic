```python
latent_dim=2
latent_inputs = keras.Input(shape=(latent_dim,))

x = tf.keras.layers.Dense(100, activation="gelu")(latent_inputs)
x = tf.keras.layers.Dense(100, activation="gelu")(x)
x = tf.keras.layers.Dense(100, activation="gelu")(x)
x = tf.keras.layers.Dense(100, activation="gelu")(x)

decoder_outputs = layers.Dense(19, activation="linear")(x)

decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()
```

# DEFINE MODEL

```python
### COSINE ANNEALING

initial_lr = 1.e-4
decay_steps = int(1e5)  # total number of steps to decay over
alpha = 1.e-1          # minimum learning rate = alpha * initial_lr

lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=initial_lr,
    decay_steps=decay_steps,
    alpha=alpha
)
```

```python
class Decoder(keras.Model):
    def __init__(self, decoder, **kwargs):
        super().__init__(**kwargs)
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss") 
        self.chi2_loss_tracker = keras.metrics.Mean(name="chi2_loss")
        self.mse_tracker = keras.metrics.Mean(name="mse")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.chi2_loss_tracker,
            self.mse_tracker,
        ]

    def train_step(self, args):
        (embedding, original, errors), = args
        with tf.GradientTape() as tape:
            reconstruction = self.decoder(embedding)
            chi2_loss = tf.reduce_mean(tf.square(original - reconstruction) / (2 * (errors ** 2))) * 1000
            mse = tf.reduce_mean(tf.square(original - reconstruction)) * 1000
            total_loss = chi2_loss
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.chi2_loss_tracker.update_state(chi2_loss)
        self.mse_tracker.update_state(mse)
        return {
            "total loss": self.total_loss_tracker.result(),
            "chi2_loss": self.chi2_loss_tracker.result(),
            "mse": self.mse_tracker.result(),
        }

    def test_step(self, args):
        embedding, original, errors = args
        reconstruction = self.decoder(embedding)
        chi2_loss = tf.reduce_mean(tf.square(original - reconstruction) / (2 * (errors ** 2))) * 1000
        mse = tf.reduce_mean(tf.square(original - reconstruction)) * 1000
        total_loss = chi2_loss
        self.total_loss_tracker.update_state(total_loss)
        self.chi2_loss_tracker.update_state(chi2_loss)
        self.mse_tracker.update_state(mse)
        return {
            "total loss": self.total_loss_tracker.result(),
            "chi2_loss": self.chi2_loss_tracker.result(),
            "mse": self.mse_tracker.result(),
        }
```
